```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Set up the Chrome WebDriver
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# URL of the website to scrape
url = "https://www.chevrolet.com.mx/"

# Open the website
driver.get(url)

# Wait for 5 seconds to allow the page to load
time.sleep(5)

# Locate and click the "Vehículos" span
vehicles_link = driver.find_element(By.XPATH, "//span[@class='q-nav-primary-link-label'][normalize-space()='Vehículos']")
vehicles_link.click()

# Optional: Wait for a few seconds to ensure the new page has loaded
time.sleep(5)

print("Clicked on 'Vehículos' and navigated to the new page.")

def click_car_type(driver, car_type):
    try:
        # Locate and click the car type link
        car_link = driver.find_element(By.XPATH, f"//a[@class='q-no-link gm-filter-by-item stat-text-link js-no-outline']//span[contains(text(),'{car_type}')]")
        car_link.click()
        print(f"Clicked on '{car_type}'.")
    except Exception as e:
        print(f"An error occurred while clicking on '{car_type}': {e}")

# List of car types to click
car_types = ["Autos", "SUVs", "Pickups", "Vans & Comerciales", "Deportivos", "Eléctricos", "Próximamente"]



# Wait for the page to load before clicking car types
time.sleep(5)

# Iterate through the car types and click each one
for car_type in car_types:
    click_car_type(driver, car_type)
    time.sleep(5)  # Wait after each click to allow the page to load

# Close the WebDriver
driver.quit()
```
